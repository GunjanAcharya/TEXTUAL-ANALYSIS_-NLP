```python
# 4: Complex Word Count

def complex_word_count(text):
    words = word_tokenize(text)

    def is_complex(word):
        syllable_count = sum(1 for char in word if char in 'aeiouAEIOU')
        return syllable_count > 2

    complex_words = [word for word in words if is_complex(word)]
    return len(complex_words)

# 4: Complex Word Count
num_complex_words = complex_word_count(article2)

# Average Sentence Length
avg_sent_len = num_words / num_sentences

# Percentage of Complex words
percentage_complex_words = num_complex_words / num_words

# Fog Index
fog_Index = 0.4 * (avg_sent_len + percentage_complex_words)

# 3: Average Number of Words Per Sentence
avg_words_per_sent= num_words/num_sentences

print(f"avg_sent_len: {avg_sent_len}")
print(f"percentage_complex_words: {percentage_complex_words}")
print(f"fog_Index: {fog_Index}")
print(f"avg_words_per_sent: {avg_words_per_sent}")
print(f"num_complex_words: {num_complex_words}")
```

```python
# Creating a dataframe of by computing the target values:

output_data = []


for i in df['URL_ID'].values:
  variables=[]

  if i==df.iloc[35]['URL_ID'] or i==df.iloc[48]['URL_ID']:     # Creating an exception for index no 35 and 48 as there are no articles in the url
    output_data.append([0 for i in range(13)])                 # all the target computation will be 0

  else:
    with open(f'/content/drive/MyDrive/Job/Blackcoffer_ Internshala/Generated_text/{i}.txt', 'r') as f:
      text = f.read()
      words = word_tokenize(text)

      # 1.1 Cleaning using Stop Words Lists
      cleaned_words = [word.lower() for word in words if word.lower() not in stop_words and word.isalpha()]

      # 1.3 Extracting Derived variables
      positive_score = sum(1 for word in cleaned_words if word in positive_words)
      negative_score = sum(1 for word in cleaned_words if word in negative_words)
      polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
      subjectivity_score = (positive_score + negative_score) / (len(cleaned_words) + 0.000001)

      for j in [positive_score,negative_score,polarity_score,subjectivity_score]:
        variables.append(j)

      # 2: Analysis of Readability
      sentences = nltk.sent_tokenize(text)
      num_words = len(words)
      num_sentences = len(sentences)

      # Average Sentence Length
      avg_sent_len = num_words / num_sentences
      variables.append(avg_sent_len)

      # Percentage of Complex words
      percentage_complex_words = num_complex_words / num_words
      variables.append(percentage_complex_words)

      # Fog Index
      fog_Index = 0.4 * (avg_sent_len + percentage_complex_words)
      variables.append(fog_Index)

      # 3: Average Number of Words Per Sentence
      avg_words_per_sent= num_words/num_sentences
      variables.append(avg_words_per_sent)

      # 4: Complex Word Count
      num_complex_words = complex_word_count(text)
      variables.append(num_complex_words)

      # 5: Word Count:
      word_cnt=word_count(text)
      variables.append(word_cnt)

      # 6: Syllable Count Per Word
      syll_lst = [syllable_count(word) for word in words if word.isalpha()]
      variables.append(syll_lst)

      # 7: Personal Pronouns
      num_pronouns = count_personal_pronouns(text)
      variables.append(num_pronouns)

      # 8: Average Word Length
      avg_word_len=avg_word_length(text)
      variables.append(avg_word_len)

      output_data.append(variables)
```
